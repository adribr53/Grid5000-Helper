```python
#Analyze Send completion time

import matplotlib.pyplot as plt
import numpy as np

def plot_data_from_file(file_path, simult_req, data_size):
    x = []
    y = []
    
    with open(file_path, 'r') as file:
        for index, line in enumerate(file):
            # Remove newline character and split the line into data values
            data_values = line.strip().split()
            
            # Extract the desired data value (e.g., the first value on the line)
            if data_values:
                data = float(data_values[0])  # Convert to float if necessary
                
                x.append(index)
                y.append(data)
                # print(x,y)
    
    plt.clf()
    plt.plot(x, y)
    plt.xlabel('#Packets')
    plt.ylabel('Send completion time (us)')
    plt.title(f'Send completion time evolution for simult_req={simult_req}, data_size={data_size}')
    # plt.show()
    print(f'Mean: {np.mean(y)}')
    plt.savefig(f'plots_logs/sender_send-compl-time_{simult_req}_{data_size}.svg', bbox_inches='tight', format='svg')

# Example usage
file_path = 'logs/durations_send_receive_1_10-100_compl_time.txt'
plot_data_from_file(file_path, '1', '10-100')

file_path = 'logs/durations_send_receive_1_100-1000_compl_time.txt'
plot_data_from_file(file_path, '1', '100-1000')

file_path = 'logs/durations_send_receive_1_1000-10000_compl_time.txt'
plot_data_from_file(file_path, '1', '1000-10000')

file_path = 'logs/durations_send_receive_10_10-100_compl_time.txt'
plot_data_from_file(file_path, '10', '10-100')

file_path = 'logs/durations_send_receive_10_100-1000_compl_time.txt'
plot_data_from_file(file_path, '10', '100-1000')

file_path = 'logs/durations_send_receive_10_1000-10000_compl_time.txt'
plot_data_from_file(file_path, '10', '1000-10000')

file_path = 'logs/durations_send_receive_100_10-100_compl_time.txt'
plot_data_from_file(file_path, '100', '10-100')

file_path = 'logs/durations_send_receive_100_100-1000_compl_time.txt'
plot_data_from_file(file_path, '100', '100-1000')

file_path = 'logs/durations_send_receive_100_1000-10000_compl_time.txt'
plot_data_from_file(file_path, '100', '1000-10000')
```

```python
import matplotlib.pyplot as plt
import numpy as np


def plot_delta_from_file(push_path, pop_path, simult_req, data_size, direction):
    delta = []
    
    with open(push_path, 'r') as file:
        lines = file.readlines()
        values_push = [float(line.strip()) for line in lines]

    with open(pop_path, 'r') as file:
        lines = file.readlines()
        values_pop = [float(line.strip()) for line in lines]
        
    for i in range(len(values_push)):
        delta.append(values_pop[i] - values_push[i])
    
    plt.clf()
    plt.plot(range(len(delta)), delta)  # Line chart with markers
    plt.xlabel(f'#push-pop')
    plt.ylabel(f'Delta push-pop time (us)')
    plt.title(f'Delta push-pop time evolution for simult_req={simult_req}, data_size={data_size}\n direction={direction}')
    # plt.show()
    print(np.mean(delta))
    plt.savefig(f'plots_logs/sender_push-pop-delta_{simult_req}_{data_size}_{direction}.svg', bbox_inches='tight', format='svg')
```

```python
#Analyze Downstream to Upstream Push and Pop delta time

push_path = 'logs/pushtimes_1.txt'
pop_path = 'logs/poptimes_1.txt'
plot_delta_from_file(push_path, pop_path, '1', '10-100', 'downstream_to_upstream')

push_path = 'logs/pushtimes_10.txt'
pop_path = 'logs/poptimes_10.txt'
plot_delta_from_file(push_path, pop_path, '10', '10-100', 'downstream_to_upstream')

push_path = 'logs/pushtimes_100.txt'
pop_path = 'logs/poptimes_100.txt'
plot_delta_from_file(push_path, pop_path, '100', '10-100', 'downstream_to_upstream')

push_path = 'logs/pushtimes_100_1000-10000.txt'
pop_path = 'logs/poptimes_100_1000-10000.txt'
plot_delta_from_file(push_path, pop_path, '100', '1000-10000', 'downstream_to_upstream')
```

```python
#Analyze Upstream to Downstream Push and Pop delta time

push_path = 'logs/uptodown_pushtimes_1.txt'
pop_path = 'logs/uptodown_poptimes_1.txt'
plot_delta_from_file(push_path, pop_path, '1', '10-100', 'upstream_to_downstream')

push_path = 'logs/uptodown_pushtimes_10.txt'
pop_path = 'logs/uptodown_poptimes_10.txt'
plot_delta_from_file(push_path, pop_path, '10', '10-100', 'upstream_to_downstream')

push_path = 'logs/uptodown_pushtimes_100.txt'
pop_path = 'logs/uptodown_poptimes_100.txt'
plot_delta_from_file(push_path, pop_path, '100', '10-100', 'upstream_to_downstream')

push_path = 'logs/uptodown_pushtimes_100_1000-10000.txt'
pop_path = 'logs/uptodown_poptimes_100_1000-10000.txt'
plot_delta_from_file(push_path, pop_path, '100', '1000-10000', 'upstream_to_downstream')

```
